```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
```

```python
peanut_butter = pd.read_csv('Peanut-Butter-Data-Set_Two-Years.csv')
```

```python
peanut_butter['Customer'].value_counts()
```

```python
peanut_butter['Item'].value_counts()
```

```python
peanut_butter['ItemID'].value_counts()
```

```python
peanut_butter
```

```python
peanut_butter['Date'] = pd.to_datetime(peanut_butter['Date'],format='%m/%d/%y')
```

```python
peanut_butter
```

```python
peanut_butter['Month'] = peanut_butter['Date'].dt.month
peanut_butter['Week'] = peanut_butter['Date'].dt.week
peanut_butter['Year'] = peanut_butter['Date'].dt.year
```

```python
peanut_butter = peanut_butter.drop(columns=['Customer','Item','Date'])
```

```python
peanut_butter['ItemID'] = peanut_butter['ItemID'].astype(str)
```

### Remove dollar signs and commas

```python
peanut_butter.dtypes
```

```python
peanut_butter['RetailUnitPrice']
```

```python
peanut_butter['RetailUnitPrice'] = peanut_butter.apply(lambda x: x['RetailUnitPrice'].replace(',',''),axis=1)
peanut_butter['RetailUnitPrice'] = peanut_butter.apply(lambda x: x['RetailUnitPrice'].replace('$',''),axis=1)

peanut_butter['RetailUnitCost'] = peanut_butter.apply(lambda x: x['RetailUnitCost'].replace(',',''),axis=1)
peanut_butter['RetailUnitCost'] = peanut_butter.apply(lambda x: x['RetailUnitCost'].replace('$',''),axis=1)

peanut_butter['RetailerUnitProfit'] = peanut_butter.apply(lambda x: x['RetailerUnitProfit'].replace(',',''),axis=1)
peanut_butter['RetailerUnitProfit'] = peanut_butter.apply(lambda x: x['RetailerUnitProfit'].replace('$',''),axis=1)

peanut_butter['TotalUnitCost'] = peanut_butter.apply(lambda x: x['TotalUnitCost'].replace(',',''),axis=1)
peanut_butter['TotalUnitCost'] = peanut_butter.apply(lambda x: x['TotalUnitCost'].replace('$',''),axis=1)

peanut_butter['RetailerProfit'] = peanut_butter.apply(lambda x: x['RetailerProfit'].replace(',',''),axis=1)
peanut_butter['RetailerProfit'] = peanut_butter.apply(lambda x: x['RetailerProfit'].replace('$',''),axis=1)
```

```python
peanut_butter['RetailUnitPrice'] = peanut_butter['RetailUnitPrice'].astype(float)
peanut_butter['RetailUnitCost'] = peanut_butter['RetailUnitCost'].astype(float)
peanut_butter['RetailerUnitProfit'] = peanut_butter['RetailerUnitProfit'].astype(float)
peanut_butter['TotalUnitCost'] = peanut_butter['TotalUnitCost'].astype(float)
peanut_butter['RetailerProfit'] = peanut_butter['RetailerProfit'].astype(float)
```

```python
peanut_butter = pd.get_dummies(peanut_butter)
```

```python
peanut_butter
```

```python
peanut_butter[(peanut_butter['ACV%Promo'] > 0) & (peanut_butter['IncrementalUnits'] == 0)]
```

```python
y = peanut_butter['IncrementalUnits']
X = peanut_butter.drop(columns=['IncrementalUnits'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=1)
```

### Linear Regression

```python
linear_regr = linear_model.LinearRegression()
```

```python
linear_regr.fit(X_train,y_train)
```

```python
y_pred = linear_regr.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test, y_pred)
```

```python
importance = linear_regr.coef_
```

```python
for i,v in enumerate(importance):
    print('Feature: %0str, Score: %.5f' % (X.columns[i],v))
```

### Lasso

```python
lasso_regr = linear_model.Lasso(alpha=0.05)
```

```python
lasso_regr.fit(X_train,y_train)
```

```python
y_pred = lasso_regr.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test, y_pred)
```

```python
importance = lasso_regr.coef_
```

```python
for i,v in enumerate(importance):
    print('Feature: %0str, Score: %.5f' % (X.columns[i],v))
```

### Decision Tree Regressor

```python
dt_regr = DecisionTreeRegressor(max_depth=3)
```

```python
dt_regr.fit(X_train,y_train)
```

```python
y_pred = dt_regr.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test, y_pred)
```
